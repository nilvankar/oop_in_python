* extension->.txt

## opening a file 
* write mode allow to make change in file
## open() function
* It is a built in function that takes (file,    
   mode='r')
this number of arguments first is name of the file and second is mode (read or write)
* In the below code first name of the file second is 'w' as write mode
* then write function gets called to add content in the file
* at last always neccessary to call close() function=None,)

## write mode functions
* write()--add multiple line
* writelines()--add single line

```python
f=open('sample1.txt','w')
f.write("This is a sample file")
f.close()
```

## read mode
* read()--Entire file gets read
* readlines()--read line by line

```python
f=open('sample1.txt','r')
res=f.read()
print(res)
f.close()
```

```python
f=open('sample1.txt','r')
res=f.readlines()
print(res)
f.close()
```

# with()
* In context manager terms to close() file automatic happend using with()
* so no need to call close() explicitly

```python
with open('sample2.txt','w') as f:
    f.write("neel vankar")
    
```

```python
with open('sample2.txt','r') as f:
    res=f.read()
    print(res)
```

## How to read large text files

```python
with open('sample1.txt','r') as f:
    print(f.read(10))
    print(f.read(10))
    print(f.read(10))
    print(f.read(10))
    
```

```python
big_r=[]
for i in range(1000):
    big_r.append("hello world")
with open('big_l.txt','w') as f:
    f.writelines(big_r)
    
```

```python
with open('big_l.txt','r') as f:
    print(f.read(900))
```

## small chunks

```python
with open('big_l.txt','r') as f:
    chunk_size=10
    while(len(f.read(chunk_size))>0):
        print(f.read(chunk_size),end="/")
        f.read(chunk_size)
        
```

## seek() and tell() functions
* tell return current location of cursor.
* seek can move anywhere in text.

```python
with open('big_l.txt','r') as f:
    print(f.read(10))
    print(f.tell())
```

```python
with open('big_l.txt','r') as f:
    print(f.read(10))
    f.seek(10)
    print(f.read(10))
```

## how to read binary files

```python
with open('WIN_20231024_19_28_53_Pro.jpg','r') as f:
    f.read()
```

```python
with open('WIN_20231024_19_28_53_Pro.jpg','rb') as f:
    with open('screenshot_copy.jpg','wb') as wf:
        wf.write(f.read())
```

## Pickling
* converting python object heirarchy into byte stream
## unpickling
*  converting byte stream heirarchy into python object

```python
class Person:
    def __init__(self,age,name,gender,email,password):
        self.age=age
        self.name=name
        self.gender=gender
        self.password=password
        self.email=email
   
    def __str__(self):
        return f"Hello I am{self.name}.I am {self.gender} and i am {self.age} years old.My Email is {self.email} and my password {self.password}"
p1=Person(name="Neel",age=18,gender="Male",email="vankar262@gmail.com",password="Neel0109*_")
print(p1)
```

```python
import pickle as pkl
with open('person.pkl','wb') as f:
    pkl.dump(p1,f)
```

```python
with open('person.pkl','rb') as f:
    print(pkl.load(f))
```
